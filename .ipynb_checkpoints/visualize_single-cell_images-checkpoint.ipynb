### Load a subset of the dataset and visualize the single-cell images

```python
import numpy
import matplotlib.pyplot as plt
import numpy as np
import os
import random
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 16))
img = plt.imread('Week1_22123/Week1_150607_B02_s1_w23FDB0AC4-EA74-4D33-A7D4-8FFC4C9ED7C8.tif')
ax.imshow(img)
ax.axis('off')
```

```python
sdir = r'Week1_22123/'

#plot a few single-cell examples
fig, axarr = plt.subplots(2, 2, figsize=(16, 16))

for i, ax in enumerate(axarr.flat):
    flist = os.listdir(sdir)
    test_img = random.choice(flist)
    test_img = os.path.join(sdir, test_img)
    img = plt.imread(test_img)
    img = numpy.array(img)
    ax.imshow(img)
    ax.axis('off')
    
plt.suptitle('Stained single-cell images')
plt.show()
```
